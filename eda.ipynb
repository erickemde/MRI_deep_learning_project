```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
neiss_aip = pd.read_csv("neiss-aip2023.tsv", sep="\t", dtype={0: str, 'icause_c': str, 'intent_c': str, 'pcause_c': str})
neiss_aip.rename(columns={"nek": "ID"}, inplace=True)
neiss_aip.head()
```

```python
neiss = pd.read_csv("neiss2023.tsv", sep="\t", dtype={0: str})
neiss.rename(columns={"CPSC_Case_Number": "ID"}, inplace=True)
neiss.head()
```

```python
merged_df = pd.merge(neiss, neiss_aip, on='ID', how='inner')
```
